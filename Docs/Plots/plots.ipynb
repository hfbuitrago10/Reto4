```python
import matplotlib.pyplot as plt
```

# **Tiempo de ejecución**
Para la recolección de datos se promedió el tiempo de ejecución para cada requerimiento después de ser ejecutado 5 veces. Las siguientes gráficas muestran los resultados de las pruebas.

```python
x = [1, 2, 3, 4, 5, 6]

y_1 = [7231.251, 965.693, 217.302, 982.437, 1561.942, 173.321]

x_labels = ["Carga de\ndatos", "1", "2", "3", "4", "5"]

plt.plot(x, y_1, marker='.')

plt.xticks(x, x_labels)
plt.xlim([0.5,6.5])
plt.xlabel('Requerimiento')
plt.ylabel('Tiempo [ms]')
plt.title('Tiempo de ejecución')

plt.show()
```
